# program to convert  text-to-speech 

# gTTS
gTTS, which stands for Google Text-to-Speech, is a Python library and command-line tool that allows you to convert text to spoken audio. It interfaces with Google's Text-to-Speech API and is useful for various applications

# IPython.display import Audio
The IPython.display module's Audio class is useful for embedding and playing audio directly

# lang='hi',tld='com'
If you want to use Google Text-to-Speech (gTTS) to convert text to Hindi and use a specific top-level domain (TLD) for the speech, you can specify the language and TLD



```python
from gtts import gTTS
from IPython.display import Audio

text_to_speech = gTTS('''Historical Heritage about Hyderabad:
Hyderabad's historical significance is highlighted by its iconic landmarks. The Charminar, built in 1591, 
stands as a testament to the city's grand past.This monument, with its four majestic minarets,
is not just a tourist attraction but also a symbol of Hyderabad's architectural prowess. Nearby, the Golconda Fort offers
a glimpse into the grandeur of the Qutb Shahi dynasty, showcasing impressive engineering and acoustics. The Qutb Shahi Tombs 
and Chowmahalla Palace further enrich the city's historical tapestry, 
making it a haven for history enthusiasts.''',lang='hi',tld='com')

text_to_speech.save('text_to_speech_gtts.wav')
sound_file = 'text_to_speech_gtts.wav'

Audio(sound_file, autoplay= False)
```

# pytssx3
pyttsx3 is a Python library that allows you to convert text to speech without the need for an internet connection. Unlike gTTS, which relies on Google's Text-to-Speech API, pyttsx3 uses offline text-to-speech engines available on your system. 

# Initialize the Engine:
### audio = pyttsx3.init()

This line initializes the text-to-speech engine, setting up everything needed to convert text to speech.

# Set Speech Rate:
### audio.setProperty('rate', 150)

This sets the speech rate to 150 words per minute. Adjusting the rate allows you to control how fast or slow the speech is.

# Set Volume:
### audio.setProperty('volume', 0.8)

This sets the volume of the speech to 80% of the maximum volume. The volume can be adjusted between 0.0 (mute) and 1.0 (maximum volume).

# Speak the Text:
### audio.say(text)

# Save the Audio File:
### audio.save_to_file(text, 'test_male_Voice.mp3')

### audio.save_to_file(text, 'test_female_Voice.mp3')

# Wait for Completion:
### audio.runAndWait()


```python
import pyttsx3
from IPython.display import Audio

text = ('''Historical Heritage about Hyderabad:
Hyderabad's historical significance is highlighted by its iconic landmarks. The Charminar, built in 1591, 
stands as a testament to the city's grand past.This monument, with its four majestic minarets,
is not just a tourist attraction but also a symbol of Hyderabad's architectural prowess. Nearby, the Golconda Fort offers
a glimpse into the grandeur of the Qutb Shahi dynasty, showcasing impressive engineering and acoustics. The Qutb Shahi Tombs 
and Chowmahalla Palace further enrich the city's historical tapestry, 
making it a haven for history enthusiasts.''')

audio = pyttsx3.init()
audio.setProperty('rate', 150)
audio.setProperty('volume', 0.8)

# change the voices
voice = audio.getProperty('voices')

# 0 for male and 1 for female
audio.setProperty('voice', voice[0].id)      # for male voice   
#audio.setProperty('voice', voice[1].id)       # for female voice

# text-to speech conversion
audio.say(text)

# save the audio file
audio.save_to_file(text, 'test_male_Voice.mp3')
#audio.save_to_file(text, 'test_female_Voice.mp3')

audio.runAndWait()
```
